```python
users = data.drop(["Product_ID", "Product_Category_1", "Product_Category_2", "Product_Category_3", "Purchase"], axis=1)
users.info()
```

```python
products = data[["User_ID", "Product_ID", "Product_Category_1", "Product_Category_2", "Product_Category_3"]]
products.info()
```

```python
user_purchase = data[["User_ID", "Purchase"]].copy()
user_purchase = user_purchase.groupby("User_ID")["Purchase"].sum().reset_index()
user_purchase.info()
```

```python
users.to_csv("data/users.csv", index=False)
products.to_csv("data/products.csv", index=False)
user_purchase.to_csv("data/user_purchase.csv", index=False)
```
